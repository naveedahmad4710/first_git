#import libraries

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# load data set

phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.show()
```

#adding titles

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# load data set

phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.show()
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# load data set

phool = sns.load_dataset("iris")
phool

# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```
